## LINEAR REGRESSION PREDICTIVE MODEL WITHOUT GROUPING

## Split the DataSet to Train and Test Dataset

```python
# These columns are going to be considered to predict the price of the Houses
columns_1 = ['log_sqft_living', 'log_bathrooms', 'log_sqft_above','log_sqft_living15']
pdx = kc_house_data[columns_1]
pdy = kc_house_data["log_price"]
# Create the train and test data by randomly performing the data split. The random_state
# (random seed) is used for reproducible results:
x_train_1,x_test_1,y_train,y_test = train_test_split(pdx, pdy, train_size =0.7, random_state = 42)
len(pdx), len(x_train_1), len(y_train)
```

### Fit a linear model using Ordinary Least Squares (OLS)

```python
#In the following code, adding constant means creating an intercept variable. If we do not
#create an intercept, the coefficients will change accordingly:

model_1 = sm.OLS(y_train,x_train_1)

#The following code creates a model summary including R-squared, adjusted R-squared, and
#the p-value of independent variables:
full_res_1 = model_1.fit()
print ("\n \n",full_res_1.summary())
```

```python
#Calculate VIF for all individual variables. 
print ("\nVariance Inflation Factor")
cnames = x_train_1.columns
for i in np.arange(0,len(cnames)):
    xvars = list(cnames)
    yvar = xvars.pop(i)
    mod = sm.OLS(x_train_1[yvar], sm.add_constant(x_train_1[xvars]))
    res = mod.fit()
    vif = 1/(1-res.rsquared)
    print (yvar,round(vif,3))
```

### Results :
- Log_sqft_living has the highest VIF value of 5.762
- All of the variables has p-value less than 0.05
- $R^2$ and Adjusted $R^2$ is 0.999
-- which means the model can explain 99.9% of our DataSet
